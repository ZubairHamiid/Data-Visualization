# seaborn and plotting

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df= sns.load_dataset('titanic')
```

```python
df.info()
```

```python
plt.figure(figsize=(16,6))

# change the scattered point colors
palette= {'male':'purple','female':'pink'}
sns.scatterplot(data=df, x='age',y='fare',hue='sex',palette=palette)
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')

# saving a plot 
# make sure resolution is in 300 dpi or 600 dpi
plt.savefig('./outputs/titanic(age vs fare)', dpi=300)
plt.savefig('./outputs/titanic(age vs fare).svg', dpi=300)
```

```python
# srtting theme or style (look on chrome)
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
plt.figure(figsize=(16,6))

# change the scattered point colors
palette= {'male':'purple','female':'pink'}
sns.scatterplot(data=df, x='age',y='fare',hue='sex',palette=palette)
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
plt.figure(figsize=(16,6))

# change the scattered point colors
palette= {'male':'purple','female':'pink'}
sns.boxplot(data=df, x='sex',y='fare',hue='sex',palette=palette) # box plot is btw num or catlg
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
plt.figure(figsize=(16,6))

# change the scattered point colors
palette= {'First':'purple','Second':'pink','Third':'green'}
sns.boxplot(data=df, x='sex',y='age',hue='class',palette=palette) # box plot is btw num or catlg
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')

# change legend loc or title
plt.legend(loc='upper right', title= 'passenger class')
```

```python
# adding buddle ploting 
# srtting theme or style (look on chrome)
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
plt.figure(figsize=(10,6))

# change the scattered point colors
palette= {'male':'purple','female':'pink'}
sns.scatterplot(data=df, x='age',y='fare',hue='sex',size='age',palette=palette)
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')
```

```python
# adding buddle ploting 
# srtting theme or style (look on chrome)
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
plt.figure(figsize=(10,6))

# change the scattered point colors
palette= {'First':'purple','Second':'pink','Third':'green'}
sns.boxenplot(data=df, x='sex',y='age',hue='class',palette=palette)
# labels
plt.xlabel(' age of male and female')
plt.ylabel('fare of all people')
plt.title('age vs fare')

# change the x tick labesl
plt.xticks(ticks=[0,1],labels=['mard','aurat'],rotation=45)
```

```python
# making scatter plot with plotly
fig= px.scatter(df, x='age',y='fare',color='class')
fig.show()
```
