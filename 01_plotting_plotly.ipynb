# plotly Data Viz ( Part-1)

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
names= ['nimra','zubair','saliha']
heights=[5.2,5.7,5.2]
```

```python

fig= px.line(x=names,y=heights,title= 'names and with person height')

fig.show()
```

```python
fig= px.bar(x=names,y=heights,title= 'names and with person height')

fig.show()
# save the plot in png file 
fig.write_image('plotly_Bar.png')
```

```python
df= sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
fig= px.scatter(df,x='age',y='fare',title='titanic dataset plot',color='sex',size='age')
fig.show()
fig.write_image('./outputs/02_plotly_Bar.png',scale=3)
```

```python
# lets go with go objects
import plotly.graph_objects as go
# create simple plot 

fig = go.Figure(data=go.Scatter(x=[1,2,3,4],y=[10,10,20,30]))
fig.show()
```

```python
# customize the upper plot
fig.update_layout(
    title= 'line plot of dummy dara',
    xaxis=dict(title='number'),
    yaxis= dict(title='values'),
    showlegend= True
)
fig.show()
```

```python
# lets go with go objects
import plotly.graph_objects as go
# create simple plot 

fig = go.Figure(data=go.Scatter(x=df['age'],y=df['fare']))
fig.show()
```

```python
# making heat map
data= np.random.rand(10,10)
heatmap= go.Heatmap(z=data)
fig= go.Figure(data=[heatmap])
fig.show()
```
