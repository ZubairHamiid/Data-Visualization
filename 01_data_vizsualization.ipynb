```python
# import libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= sns.load_dataset('titanic')
```

```python
df.plot (kind='scatter', x='age',y='fare',color='red')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.xlabel('gender')
plt.ylabel('Avg fare')
```

```python
df.head()
```

```python
# seaborn plots

sns.barplot(data= df, x='survived', y='sex', hue='who')
```

```python
sns.barplot(data= df, x='sex', y='age', hue='class')
```

```python
sns.boxplot(data= df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data= df, x='fare', y='age', hue='class')
```

```python
#hyphothesis to check is last 2 points are male or female
sns.scatterplot(data= df, x='fare', y='age', hue='sex')
```

```python
# make scatter plot via plotly
import plotly.express as px

fig = px.scatter(df, x='age',y='fare',color='class')
fig.show()
```
