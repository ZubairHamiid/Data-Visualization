# Sun Burst Plot

```python
# import libraries
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
# sunburst plot 
df= sns.load_dataset('titanic')
```

```python
df.columns
```

```python
df.dtypes
```

```python
fig = px.sunburst(df,path=['sex', 'class', 'who','alone' ,'alive'],
                  values='survived',color='sex')

fig.show()
```
